```python
def tidyCovid():
    notSelect = time_series_covid_19_confirmed.columns.isin(["Province/State","Lat","Long"])
    Recovered = pd.melt(time_series_covid_19_confirmed.loc[:,~notSelect],
                    "Country/Region", var_name = "Date", value_name = "Recovered")
    Dead = pd.melt(time_series_covid_19_deaths.loc[:,~notSelect],
               "Country/Region", var_name = "Date", value_name = "Dead")
    Confirmed = pd.melt(time_series_covid_19_confirmed.loc[:,~notSelect],
              "Country/Region", var_name = "Date", value_name = "Confirmed")
    Dead.sort_values(by = "Dead", ascending = False).head()
    Recovered.sort_values(by = "Recovered", ascending = False).head()
    print(Confirmed.shape)
    Confirmed.sort_values(by = "Confirmed", ascending = False).head()
    covid = pd.concat([Confirmed, Dead.iloc[:,-1], Recovered.iloc[:,-1]], axis = 1)
    return covid

tidy = tidyCovid()
tidy
```
